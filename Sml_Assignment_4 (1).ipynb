Read the data with pandas and find features and target variables

```python
import pandas as pd
d=pd.read_csv("/train.csv")
d
```

```python
y=d['price_range']
y
```

```python
x=d.drop('price_range',axis=1)
x
```

Normalize the data with min-max scaling

```python
df=x-x.min()/(x.max()-x.min())
df
```

Split the data into train and test.

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df,y,test_size=0.2,random_state=40)
x_test,x_train,y_test,y_train
```
